```python
f = open(path + "/newfile.txt","r")

p = 0
fa= 0

for i in f:
    # Strip newline characters and split by comma
    parts = i.strip().split(',')
    # Ensure there are at least two parts before accessing indices
    if len(parts) >= 2:
        name = parts[0]
        marks = int(parts[1])
        if marks > 45:
            print(name)
            p = p + 1
        else:
            print(name)
            fa = fa + 1
    else:
        print(f"Skipping malformed line: {i.strip()}")
print("Passed =", p)
print("Failed =", fa)
f.close()
```
