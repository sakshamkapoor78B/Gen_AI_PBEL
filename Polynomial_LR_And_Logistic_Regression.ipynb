```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

full_health_data = pd.read_csv("/content/drive/MyDrive/Gen_AI_PBEL/CSV/data29Jun.csv")
full_health_data = full_health_data.dropna()

x = pd.to_numeric(full_health_data["Average_Pulse"])
y = pd.to_numeric(full_health_data["Calorie_Burnage"])

slope, intercept, r, p, std_err = stats.linregress(x, y)
# slope = Slope (m) of the regression line
# intercept = Intercept (c)
# r = Correlation Coefficient
# p= P-value
# std_err = standard error

plt.scatter(x,y)
plt.plot(x, intercept + slope*x, 'r', label='Regression Line')
plt.xlabel("Average_Pulse")
plt.ylabel("Calorie_Burnage")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)
# slope = Slope (m) of the regression line
# intercept = Intercept (c)
# r = Correlation Coefficient
# p= P-value
# std_err = standard error

def myfunc(x):
  return slope* x+ intercept

mymodel = list(map(myfunc,x))

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
```

# Logistic Regression

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Already present
df= sns.load_dataset("iris")
df
```

```python
df.head()
```

```python
df1 = df [df["species"]!="setosa"]
df1
```

```python
df1["species"].unique()
```

```python
# species are in object so convert species into categorical
# converting data using mapping
df1["species"] = df1["species"].map({"versicolor" : 0 , "virginica" : 1})
df1
```

```python
df1["species"].unique()
```

```python
# DIvide the data in input and output
x = df1.iloc[:,:-1]
y = df1.iloc[:,-1]

x.head(2)
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.25 , random_state = 42)
```

```python
x_train.head()
```

```python
y_train.head()
```

# model building


```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(x_train,y_train)                  # 1/1+e^-z ==>   z =m1x1+m2x2+m3x3+m4x4+c
```

```python
y_pred = logistic_model.predict(x_test)
```

```python
y_test ,y_pred
```

```python
x_pred = logistic_model.predict(x_train)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
print("Training Accuracy :" , accuracy_score(y_train,x_pred))
print("Testing Accuracy :" , accuracy_score(y_test,y_pred))
```

```python
print (classification_report(y_test,y_pred))
```
