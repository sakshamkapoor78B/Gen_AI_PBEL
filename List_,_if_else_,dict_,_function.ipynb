```python
# Arguments Types
# 1. Default Argument
# 2. Positional Argument
# 3. Keyword Argument
# 4. Arbitary Argument : Variable lenght: Positional, Keyword
```

```python
#  Positional Argument
def add(a,b):
  return a+b
```

```python
add(45,89)
```

```python
# Defalut
def add(a, b=90):
  return a+b
```

```python
add(67,78)
```

```python
# Keyword
def add(a,b):
  return a+b
add(a= 23 , b = 89)
```

```python
# Arbitary Argument
def data(*a):
  return a
```

```python
data (12,13,1,5,45,78,85)
```

```python
def data(b,*a):
  return a,b
```

```python
def data(**a):
  return a
```

```python
data (b = 78 , c=67, d =98, e=76)
```
