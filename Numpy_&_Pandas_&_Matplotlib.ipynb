```python
import numpy as np
weather = {
    "temp": [23,24,55,12,34,90],
    "humi": [23,22,12,11,45, np.nan],
    "wind": [67,89,90,34,22, np.nan],
    "event": ["sunny", "rainy","sunny","snow","rainy", np.nan]
}
```

```python
df1 = pd.DataFrame(weather)
df1
```

```python
df1.isnull()
```

```python
df1.isnull().sum()
```

```python
# fillna(0/mean/median) , dropna(axis=0/1) , inplace (actual dataframe)
```

```python
df1.fillna(0)
```

```python
df1.dropna(axis = 1)                   # axis => 0/1
```

```python
df1.dropna(axis = 0, inplace = True)
df1
```

```python
df1["event"]=df1["event"].map({"sunny" :0 , "rainy":1 ,"snow":2})
df1
```
