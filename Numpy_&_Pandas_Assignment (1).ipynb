```python
df = pd.read_csv("/content/drive/MyDrive/IBM/CSV/Day,Temperature.csv")
df
```

```python
highest_temp = np.max(df["Temperature"])
print("Highest Temp :", highest_temp)
lowest_temp = np.min(df["Temperature"])
print("lowest Temp :", lowest_temp)
Average_Temp = np.mean(df["Temperature"])
print ("Average Temp :", Average_Temp)
```

```python
def classify(temp):
  if temp < 20:
    print ("cold")
  elif temp > 30:
    print("hot")
  else:
    print ("moderate")
print("\nTemperature Categories:")
for i in range(len(df)):
    day = df["Day"][i]
    temp = df["Temperature"][i]
    category = classify(temp)
    print(day, "-", temp, "°C :", category)
```
