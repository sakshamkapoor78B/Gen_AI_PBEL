using functions to calculate the total marks, percentage, and final grade.

```python
def calc_total(marks):
  return sum(marks)
def calc_percentage(total,number_subjects):
  return total/number_subjects
def assign_grade(percentage):
  if percentage >= 90:
    print ("A")
  elif percentage >= 75:
    print ("B")
  elif percentage >= 60:
    print ("C")
  elif percentage >= 40:
    print ("D")
  else:
    print ("E")

marks = []
for i in range(5):
  mark = float(input(f"Enter marks for Subject {i+1}: "))
  marks.append(mark)

total = calc_total(marks)
print(total)

Percentage = calc_percentage(total ,5)
print(Percentage,"%")

final= assign_grade(Percentage)
print(final)
```

#Practical 1: Student Performance Management System

A school wants to automate the process of managing student records. Develop a Python program
that stores student details and marks, calculates percentages, assigns grades, and displays the
topper.

```python
grade_criteria = (("A", 90),("B", 75),("C", 60),("D", 40),("F", 0))

def calculate_total(marks):
    return sum(marks)

def calculate_percentage(total):
    return total / 5

def assign_grade(percentage):
    if percentage >= 90:
        return "A"
    elif percentage >= 75:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 40:
        return "D"
    else:
        return "F"

def display_student(student):
    print("\nRoll No:", student["roll_no"])
    print("Name:", student["name"])
    print("Marks:", student["marks"])
    print("Total:", student["total"])
    print("Percentage:", round(student["percentage"], 2), "%")
    print("Grade:", student["grade"])

students = []
n = int(input("Enter number of students: "))

for i in range(n):
    print(f"\nEnter details for Student {i+1}")

    roll_no = input("Enter Roll Number: ")
    name = input("Enter Name: ")

marks = []
for j in range(5):
        mark = float(input(f"Enter marks of Subject {j+1}: "))
        marks.append(mark)

total = calculate_total(marks)
percentage = calculate_percentage(total)
grade = assign_grade(percentage)

student = {
        "roll_no": roll_no,
        "name": name,
        "marks": marks,
        "total": total,
        "percentage": percentage,
        "grade": grade
    }

students.append(student)

for student in students:
    display_student(student)

topper = students[0]

for student in students:
    if student["percentage"] > topper["percentage"]:
        topper = student

print("\nTopper")
print("Name:", topper["name"])
print("Roll No:", topper["roll_no"])
print("Percentage:", round(topper["percentage"], 2), "%")
print("Grade:", topper["grade"])
```

Q3. A file (attendance.txt) contains the attendance status of students
(Present/Absent). Read the file and display only the names of absent students. Also
calculate the total number of present and absent students.

```python
file = open("/content/drive/MyDrive/IBM/attendance.txt","r")
file.close()
```

```python
path = ("/content/drive/MyDrive/IBM")
file1 = open(path+"/attendance.txt","w+")
file1.write("Saksham,Present \nVarun,Present \nNitin,Absent \nSuhesh,Present \nNaina,Absent \nManish,Absent \nAmit,Present")
file1.seek(0)
content = file1.read()
print(content)
file1.close()
```

```python
with open(path+"/attendance.txt","r") as file:
  for line in file:
    data = line.strip().split(",")
    print(data[1])
```

```python
f = open (path+"/attendance.txt","r")
present = 0
absent = 0

for i in f:
    name, status = i.strip().split(',')

    if status.strip() == "Absent":
        print("Absent: ", name)
        absent = absent + 1
    else:
        present = present + 1

print("Total Present =", present)
print("Total Absent =", absent)

f.close()
```
